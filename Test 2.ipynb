#### split into training and validation data set

```python
# split data into training and validation data sets

# separate A's and B's
ix_A = (labels==0)
ix_B = (labels==1)
images_A = images[ix_A,:,:]
labels_A = labels[ix_A]
images_B = images[ix_B,:,:]
labels_B = labels[ix_B]

num_train_examples = 70000

# concatenate A and B into training and validation data sets
images_train = np.vstack([images_A[:int(num_train_examples/2),:,:],
                          images_B[:int(num_train_examples/2),:,:]])
labels_train = np.hstack([labels_A[:int(num_train_examples/2)],
                          labels_B[:int(num_train_examples/2)]])

images_valid = np.vstack([images_A[int(num_train_examples/2):,:,:],
                          images_B[int(num_train_examples/2):,:,:]])
labels_valid = np.hstack([labels_A[int(num_train_examples/2):],
                          labels_B[int(num_train_examples/2):]])

# scramble data sets
ix = np.random.permutation(num_train_examples)
images_train = images_train[ix,:,:]
labels_train = labels_train[ix]
ix = np.random.permutation(100000-num_train_examples)
images_valid = images_valid[ix,:,:]
labels_valid = labels_valid[ix]

print('training')
print(images_train.shape)
print(labels_train.shape)
print('validation')
print(images_valid.shape)
print(labels_valid.shape)
```

#### resize and scale data

```python
print("max pixel value = %f" % np.max(images_train[:, :, :]))
print("min pixel value = %f" % np.min(images_train[:, :, :]))
print("mean pixel value = %f" % np.mean(images_train[:, :, :]))
print("std pixel value = %f" % np.std(images_train[:, :, :]))
```

```python
# reshape image tensors to a matrix 
image_shape = 28*28
inputs_train = images_train.reshape(-1,image_shape)
inputs_valid = images_valid.reshape(-1,image_shape)
inputs_test = images_test.reshape(-1,image_shape)
```

```python
# scale data
inputs_train = scale(inputs_train.astype('float64'),axis=1)
inputs_valid = scale(inputs_valid.astype('float64'),axis=1)
inputs_test = scale(inputs_test.astype('float64'), axis=1)

inputs_train = np.array(inputs_train.astype('float64'))
inputs_valid = np.array(inputs_valid.astype('float64'))
inputs_test = np.array(inputs_test.astype('float64'))
```

#### one-hot encode labels

```python
# one-hot-encode labels
outputs_train = OneHotEncoder(sparse=False).fit_transform(labels_train.reshape(-1,1))
outputs_valid = OneHotEncoder(sparse=False).fit_transform(labels_valid.reshape(-1,1))
outputs_test = OneHotEncoder(sparse=False).fit_transform(labels_test.reshape(-1,1))

outputs_train = np.array(outputs_train.astype('float64'))
outputs_valid = np.array(outputs_valid.astype('float64'))
outputs_test = np.array(outputs_test.astype('float64'))

print('Data reshaped, scaled and one-hot-encoded:')
print('   training input:',inputs_train.shape,' output:',outputs_train.shape)
print(' validation input:',inputs_valid.shape,' output:',outputs_valid.shape)
```

#### minibatch function used for stochastic gradient descent

```python
def minibatches(x, y, batchsize):
    assert x.shape[0] == y.shape[0]
    indices = np.arange(x.shape[0])
    np.random.shuffle(indices)
    
    for i in range(0, x.shape[0] - batchsize + 1, batchsize):
        excerpt = indices[i:i + batchsize]
        yield np.array(x[excerpt], dtype='float64'), np.array(y[excerpt], dtype='float64')
    
    leftOver = (x.shape[0] - batchsize) % batchsize
    if leftOver != 0:
        excerpt = indices[len(indices) - leftOver:]
        yield np.array(x[excerpt], dtype='float64'), np.array(y[excerpt], dtype='float64')
```

#### Establish baseline accuracy rate using logistic regression

```python
# inputs and expected output
x = tf.placeholder(tf.float32, shape=(None, image_shape))
y = tf.placeholder(tf.float32, shape=(None, 2))

# input --> output
W = tf.Variable(tf.truncated_normal([image_shape, 2], dtype=tf.float32, seed=0, stddev=0.01))
b = tf.Variable(tf.truncated_normal([2], dtype=tf.float32, seed=0, stddev=0.01))
logits = tf.matmul(x, W) + b
y_pred = tf.argmax(tf.nn.softmax(logits), dimension=1)

# loss function and optimizer
CE = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, y))
optimizer = tf.train.AdamOptimizer().minimize(CE)

# error rate
accuracy = 1.0 - tf.contrib.metrics.accuracy(y_pred, tf.argmax(y, dimension=1))
```

```python
init = tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(init)
    num_epochs = 0
    starting_time = time.time()

    while num_epochs < 10:

        for batch in minibatches(inputs_train, outputs_train, 1000):
            batch_x, batch_y = batch
            _ = sess.run([optimizer], feed_dict={x: batch_x, y: batch_y})

        num_epochs += 1

        training_CE, training_error = sess.run([CE, accuracy],
                                                  feed_dict={x: inputs_train,
                                                             y: outputs_train})
        validation_CE, validation_error = sess.run([CE, accuracy],
                                                  feed_dict={x: inputs_valid,
                                                             y: outputs_valid})

        print("EPOCH = {:d}".format(num_epochs))
        print("    training cross-entropy = {:.5f}".format(training_CE))
        print("  validation cross-entropy = {:.5f}".format(validation_CE))
        print("       training error rate = {:.3f}".format(training_error))
        print("     validation error rate = {:.3f}".format(validation_error))
        print("      total time (minutes) = {:.1f}".format((time.time() - starting_time) / 60.0))
```

#### Set hyper-parameters

```python
# hyper-parameters
EPOCHS = [10, 100, 1000, 10000]                    # number of training epochs 
N1     = [100, 1000, 10000, 50000, 70000]          # training set size
NODES  = [128, 256, 512, 1024, 2048, 4096]         # nodes in hidden layer
LAMBDA = [0, 0.0001, 0.001, 0.01, 0.1, 1.0, 5.0]   # regularization parameter
BS     = [10, 100, 1000, 5000, 10000]              # batch size
STD    = [0.001, 0.01, 0.1, 1.0, 5.0]              # weight initialization standard deviation
ALPHA  = [0.0001, 0.001, 0.01, 0.1, 1.0, 5.0]      # learning rate
```

```python
epochs     = 12
n1         = N1[4]
nodes      = NODES[2]
lambda_    = LAMBDA[2]
batch_size = BS[2]
std        = STD[1]
alpha      = ALPHA[1]
early_stopping = 5    # how many epochs of no valid error improvement
```

#### define Tensorflow computational graph

```python
# inputs and expected output
x = tf.placeholder(tf.float32, shape=(None, image_shape))
y = tf.placeholder(tf.float32, shape=(None, 2))

# input --> hidden layer
W1 = tf.Variable(tf.truncated_normal([image_shape, nodes], dtype=tf.float32, seed=0, stddev=std))
b1 = tf.Variable(tf.truncated_normal([nodes], dtype=tf.float32, seed=0, stddev=std))
y1 = tf.nn.relu(tf.matmul(x, W1) + b1)

# hidden layer --> output
W2 = tf.Variable(tf.truncated_normal([nodes, 2], dtype=tf.float32, seed=0, stddev=std))
b2 = tf.Variable(tf.truncated_normal([2], dtype=tf.float32, seed=0, stddev=std))
logits = tf.matmul(y1, W2) + b2
y_pred = tf.argmax(tf.nn.softmax(logits), dimension=1)

# loss function and optimizer
l2 = lambda_ * tf.nn.l2_loss(W1) + lambda_ * tf.nn.l2_loss(W2)
CE = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, y) + l2)
optimizer = tf.train.AdamOptimizer(learning_rate=alpha).minimize(CE)

# error rate
accuracy = 1.0 - tf.contrib.metrics.accuracy(y_pred, tf.argmax(y, dimension=1))
```

#### train network

```python
init = tf.initialize_all_variables()
train_error = []
valid_error = []
train_CE = []
valid_CE = []

sess = tf.Session()
sess.run(init)

num_epochs = 0
starting_time = time.time()

#     last_valid_error_rate = 1.0
#     num_times_not_improved = 0

while num_epochs < epochs:

#         if num_times_not_improved >= early_stopping:
#             break

    for batch in minibatches(inputs_train[:n1], outputs_train[:n1], batch_size):
        batch_x, batch_y = batch
        _ = sess.run([optimizer], feed_dict={x: batch_x, y: batch_y})

    num_epochs += 1

    training_CE, training_error, l2_loss = sess.run([CE, accuracy, l2],
                                              feed_dict={x: inputs_train,
                                                         y: outputs_train})
    validation_CE, validation_error = sess.run([CE, accuracy],
                                              feed_dict={x: inputs_valid,
                                                         y: outputs_valid})
    
#         if validation_error > last_valid_error_rate:
#             num_times_not_improved += 1
#         else:
#             num_times_not_improved = 0
#         last_valid_error_rate = validation_error

    # store status of training
    train_error = train_error[:] + [training_error]
    valid_error = valid_error[:] + [validation_error]
    train_CE = train_CE[:] + [training_CE]
    valid_CE = valid_CE[:] + [validation_CE]

    print("EPOCH = {:d}".format(num_epochs))
    print("    training cross-entropy = {:.5f}".format(training_CE))
    print("  validation cross-entropy = {:.5f}".format(validation_CE))
    print("       training error rate = {:.3f}".format(training_error))
    print("     validation error rate = {:.3f}".format(validation_error))
    print("                   l2 loss = {:.5f}".format(l2_loss))
    print("      total time (minutes) = {:.1f}".format((time.time() - starting_time) / 60.0))
```

```python
plt.plot(train_error, color="blue", label="training error")
plt.plot(valid_error, color="green", label="validation error")
plt.plot(valid_CE, color="red", label="validation CE")
plt.plot(train_CE, color="yellow", label="training CE")
plt.legend(loc='upper center')
plt.show()
```
