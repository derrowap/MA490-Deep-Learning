```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Part (a)

```python
MNIST = np.load('./Data/MNIST_5.npz')
MNIST.keys()
```

```python
images = MNIST['images']
images = images.reshape((5, 28, 28)) # Fortran order
images.shape
```

```python
plt.figure(figsize=(1,1))
plt.axis('off')
for i in range(len(images)):
    plt.imshow(images[i], cmap="Greys_r")
    print("LABEL: %s" % MNIST['labels'][i])
    plt.show()
```

```python
images[0][0]
```

The labels do line up

Rank: 2, Shape: (5, 28, 28), Type: float32

#### Part (b)

```python
RGB = np.load('./Data/CIFAR-10_select_5.npz')
RGB.keys()
```

```python
images = RGB['images']
images.shape
```

```python
images = images.reshape((5, 32, 32, 3), order="F")
images.shape
```

```python
plt.figure(figsize=(1,1))
plt.axis('off')
for i in range(len(images)):
    image = images[i].transpose((1, 0, 2))
    plt.imshow(image, cmap="Greys_r")
    print("LABEL: %s" % RGB['label_names'][RGB['labels'][i]])
    plt.show()
```

```python
images[0][0][0]
```

The labels do line up.

Rank: 4, shape: (5, 32, 32, 3), type: uint8
