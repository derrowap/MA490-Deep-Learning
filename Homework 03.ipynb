#### Part (a): How does the RMSE for you linear model compare to the standard deviation of the LogRatios?

```python
df = pd.read_csv('./Data/Stocks15yr.csv')
df.head()
```

```python
# conert dates to date-time format
df['Date'] = pd.to_datetime(df.Date)
df.dtypes
```

```python
# pivot closing prices
df = df.pivot(index='Date', columns='ticker', values='Close')
df.head(12)
```

```python
df.shape
```

```python
# check for bad/missing values
df.isnull().sum().sum()
```

```python
# forward fill and black fill missing prices
df.fillna(method='ffill', axis=0).fillna(method='bfill', axis=0).isnull().sum().sum()
```

```python
df = df.fillna(method='ffill', axis=0).fillna(method='bfill', axis=0)
df.head(13)
```

```python
# compute logs of closing prices
df1 = np.log(df)
df1.head()
```

```python
# shit log price forward one day
df1.shift().head()
```

```python
# compute LogRatios
df2 = df1 - df1.shift()
df2 = df2.dropna()
df2.head(15)
```

```python
# plot Google's LogRatios
df2.GOOG.plot()
```

```python
# look for stock splits
print('Google minimum LogRatio', df2.GOOG.min())
print('Google minimum price ratio', np.exp(df2.GOOG.min()))
```

This looks like a stock split.

```python
# compute standard deviation for all LogRatios
std = df2.values.std()
num_deviations = 5
```

```python
# check for extreme variations
(df2.abs() > num_deviations * std).sum().sum()
```

```python
# remove extreme variations
# FOR HOMEWORK: change this to "3*std", or 3 standard deviations away because 0.1 throws away too much data
df2[df2.abs() > num_deviations * std] = 0

# check for extreme variations
(df2.abs() > num_deviations * std).sum().sum()
```

```python
# define inputs and outputs
inputs = df2.values[0:-1, :]
outputs = df2.values[1:,:]

print("Shapes:")
print(inputs.shape)
print(outputs.shape)

# today's values are used to predict tomorrow's value. So the data is shifted
print("\nSample values:")
print(inputs[0:3, 0:3])
print(outputs[0:3, 0:3])
```

```python
# Tensorflow

# y = wx + b
x = tf.constant(inputs, dtype='float32', shape=[4171, 467])
y = tf.constant(outputs, dtype='float32', shape=[4171, 467])

w = tf.Variable(tf.truncated_normal([467, 467], stddev = 0.0001))
b = tf.Variable(tf.truncated_normal([467], stddev=0.001))

# error loss function
MSE = tf.reduce_mean(tf.square(tf.matmul(x, w) + b - y))

# define optimizer
optimizer = tf.train.AdamOptimizer(learning_rate=0.01, epsilon=1e-10).minimize(MSE)

# prediction with current weights
y_pred = tf.matmul(x, w) + b

# init all variable
init = tf.initialize_all_variables()
```

```python
sess = tf.Session()
sess.run(init)
```

```python
MAXSTEPS = 500
for step in range(MAXSTEPS + 1):
    (_, mse, w0, b0) = sess.run([optimizer, MSE, w, b])
    if (step % 100) == 0:
        print('step = %-5d MSE = %-10f' % (step, mse))
print('done!')
```

```python
# make predictions
predictions = sess.run(y_pred)
```

```python
# check mean square error
mse = ((outputs - predictions) ** 2).mean()
rmse = np.sqrt(mse)
print('MSE = ', mse)
print('RMSE = ', rmse)
print('output\'s standard deviation = ', outputs.std())
print('difference between RMSE of model and std of LogRatios: ', (outputs.std() - rmse))
```
