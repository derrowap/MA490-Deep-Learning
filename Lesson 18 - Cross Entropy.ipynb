```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# cross entropy H(p, q) between two discrete probability distributions p and q
def H(p, q):
    return (p * np.log2(p / q)).sum()

# mean squared error
def MSE(u, v):
    return ((u - v) ** 2).mean()

# softmax
def softmax(x):
    epsilon = 1e-10
    return np.exp(x / (x.max() + epsilon)) / np.exp(x / (x.max() + epsilon)).sum()
```

```python
A = 100
n = 100
MSE_ = []
CE_ = []
for i in range(1000):
    p = softmax(A * np.random.rand(n))
    q = softmax(A * np.random.rand(n))
    MSE_ = MSE_[:] + [MSE(p, q)]
    CE_ = CE_[:] + [H(p, q)]
```

```python
mse = np.array(MSE_)
ce = np.array(CE_)
plt.scatter(x=mse, y=ce, s=50, alpha=0.1)
plt.xlabel('MSE')
plt.ylabel('CE')
plt.xlim(xmin=mse.min(), xmax=mse.max())
plt.ylim(ymin=ce.min(), ymax=ce.max())
```
