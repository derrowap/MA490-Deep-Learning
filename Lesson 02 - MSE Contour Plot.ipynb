# Lesson 02 - MSE Contour Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get data
df = pd.read_csv('../Data/height_weight_10.csv')
print(df)
```

```python
# define functions
def MSE(u, v):
    return ((u - v) ** 2).mean()
def f(x, a, b):
    return a * x + b
```

```python
# draw countour lines
a = np.linspace(2, 10, 100) # get 100 equally spaced values between 5 and 10
b = np.linspace(-500, -100, 100)

(aa, bb) = np.meshgrid(a, b) # make each a square matrix

x = df.height
y = df.weight

mse = np.zeros((100, 100)) # init mse to same size as aa and bb matrices

for i in range(100):
    for j in range(100):
        mse[i, j] = MSE(f(x, aa[i, j], bb[i, j]), y)
plt.contour(aa, bb, mse, 25)
```
